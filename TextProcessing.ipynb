```python
import numpy as np 
import pandas as pd
import math 
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import glob
import os
import string
import sys
import random
import math
from stemming.porter2 import stem
stopwords_f = open('common-english-words.txt', 'r') 
stop= stopwords_f.read().split(',')

def clean_df(df):
    query_dict={}
    for i in range(len(df['text'])): 
        u={}

        temp=df['text'][i].strip()
        temp=temp.split()

        for term in temp: 
            term = term.translate(str.maketrans('','', string.digits)).translate(str.maketrans(string.punctuation, ' '*len(string.punctuation)))
            term = term.replace("\\s+", " ")
            term = term.strip()
            term = stem(term.lower())

            if len(term) > 2 and term not in stop:
                try:
                    u[term] += 1/len(temp)
                except KeyError:
                    u[term] = 1/len(temp)
        query_dict[df['index'][i]]=u
    return query_dict
```

```python
df=pd.read_csv('AMI_train.csv')
df.reset_index(inplace=True)
```

```python
clean_df(df)
```
